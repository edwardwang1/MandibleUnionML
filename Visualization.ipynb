```python
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
def normalize(volume):
    """Normalize the volume"""
    min = 200
    max = 2200
    volume[volume < min] = min
    volume[volume > max] = min
    volume = (volume - min) / (max - min)
    volume = volume.astype("float32")
    return volume
```
